## Loading data

Loading data, settings for graphs and essential libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
plt.rcParams.update({'font.size': 15, 'figure.figsize': (15, 15)})

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv('https://raw.githubusercontent.com/dbonacorsi/AMLBas2122/main/datasets/pima-indians-diabetes.data.csv',
                 names=names)
```

## Useful Functions

Simple first description of the dataset

## Barplots

Barplots for each features, separated per class

```python
fig, axes = plt.subplots(3, 3)

for ax, col in zip(axes.ravel(), names):
    sns.histplot(data=df, x=col, hue='class', ax=ax, multiple='stack', palette='mako')
    
    ax.set_xlabel(col)

plt.tight_layout()
plt.show()
```

## Correlation Matrix

Correlation Matrix with visualization, annotations and a nice cmap

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

corrs = df.corr()

sns.heatmap(corrs, annot=True, vmin=-1, vmax=1, cmap='coolwarm')

plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

## Scatters plots

Scatters of each feature against each other, separated by class. It takes a while with a lot of features

```python
sns.pairplot(df, hue='class', palette='mako')

plt.tight_layout()
plt.show()
```

## Boxplots

Simple Boxplots 

```python
df.plot(kind='box', subplots=True, layout=(3, 3))

plt.tight_layout()
plt.show()
```
